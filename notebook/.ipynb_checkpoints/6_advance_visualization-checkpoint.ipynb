```python
new_df = main[['game_id','action_team_name','home_team','away_team','home_team_side','away_team_side','coordinates']]
```

```python
new_df
```

```python
df = new_df[new_df["action_team_name"] == new_df["home_team"]][["action_team_name",'game_id',"home_team_side",'coordinates']]
df = df.rename(columns={"home_team_side":"side"})
df
```

```python
df1 = new_df[new_df["action_team_name"] == new_df["away_team"]][["action_team_name",'game_id',"away_team_side",'coordinates']]
df1 = df1.rename(columns={"away_team_side":"side"})
df1
```

```python
df = pd.concat([df,df1],axis=0)
```

```python
final_df = df[df["side"] != "Side Not Available"]
final_df
```

```python
def get_offense_corr(row):
    
    team_side = row.loc["side"]
    corr = row.loc["coordinates"]#[1:-1].split(", ")
    if corr[0] != "NA":
        x = float(corr[0])
        if corr[1] != "NA":
            y = float(corr[1])
            
            if team_side == 'right':
                return (y,-x)
            elif team_side == 'left':
                return (-y,x)
        else:
            return None

            
    else:
        return None
```

```python
final_df['new_corr'] = final_df.apply(get_offense_corr,axis=1)
```
